```python
!pip install yfinance
!pip install seaborn
import seaborn as sns
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

gbtc = yf.Ticker("GBTC").history(period="max")

# Calculate the 20-day moving average of the closing price
gbtc["20d"] = gbtc["Close"].rolling(window=20).mean()

# Plot the closing price and the 20-day moving average
plt.plot(gbtc.index, gbtc["Close"], label="Close")
plt.plot(gbtc.index, gbtc["20d"], label="20-day Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("GBTC Close Price and 20-day Moving Average")
plt.legend(loc="upper left")
plt.show()

plt.bar(gbtc.index, gbtc["Volume"])
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("GBTC Daily Volume")
plt.show()

# Calculate the correlation matrix
corr = gbtc.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, cmap=plt.cm.RdYlBu_r)
plt.title("GBTC Correlation Matrix")
plt.show()
```
