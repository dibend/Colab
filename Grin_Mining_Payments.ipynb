```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime

plt.style.use("bmh")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200

url = "https://grin.2miners.com/api/accounts/grin17cha4wcp3872udu0vr894lzdpnywpdrhcynve8dqun6k0x5ex4qqgehl6s"

response = requests.get(url)

data = response.json()

amounts_and_timestamps = [(payment['amount'] / 1000000000, datetime.fromtimestamp(payment['timestamp'])) for payment in data['payments']]

# Extract the amounts and timestamps into separate lists
amounts, timestamps = zip(*amounts_and_timestamps)

# Plot the amounts against the timestamps
plt.scatter(timestamps, amounts, c='gold', edgecolors='black')

# Add a title and labels to the axes
plt.title("Payment amounts over time")
plt.xlabel("Timestamp")
plt.ylabel("Amount")

# Show the plot
plt.show()
```
