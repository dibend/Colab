The **Zillow Home Value Forecast** (ZHVF) is the month-ahead, quarter-ahead and year-ahead forecast of the Zillow Home Values Index (ZHVI). ZHVF is created using the all homes, mid-tier cut of ZHVI and is available both raw and smoothed, seasonally adjusted.

[Zillow Data Source](https://www.zillow.com/research/data/)

```python
 #import packages
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("bmh")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200

#load data
df = pd.read_csv("https://files.zillowstatic.com/research/public_csvs/zhvf_growth/Zip_zhvf_growth_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv?t=1673388230")

#user input
zip_code = input("Enter a zip code: ")

#filter data
df_zip = df[df['RegionName'] == int(zip_code)]

#plot data
plt.figure(figsize=(10, 4))
plt.plot(df_zip.iloc[0, 10:].index.values, df_zip.iloc[0, 10:].values, marker='o')
plt.title("Price Change Predictions for Zip Code " + zip_code)
plt.xlabel("Date")
plt.ylabel("Predicted Price Change (%)")
plt.show()
```
