```python
!pip install yfinance plotly

import yfinance as yf
import plotly.graph_objects as go
from datetime import datetime
import time
from IPython.display import display, clear_output

def fetch_data(stock='AAPL'):
    data = yf.download(tickers=stock, period='1d', interval='1m')
    return data

def update_plot(stock='AAPL'):
    while True:
        data = fetch_data(stock)
        clear_output(wait=True)  # Clear the previous plot
        # Create a new plot
        fig = go.Figure(data=[go.Candlestick(x=data.index,
                                             open=data['Open'],
                                             high=data['High'],
                                             low=data['Low'],
                                             close=data['Close'])])
        fig.update_layout(title=f'Live share price for {stock}: {datetime.now()}',
                          xaxis_title='Time',
                          yaxis_title='Price (USD)')
        fig.show()
        time.sleep(60)  # Update every minute

update_plot(input("Asset Ticker: "))  # Replace 'AAPL' with any ticker you want to track
```
