<h1><b>Zillow Seasonally Adjusted Smoothed Home Value Data for Zip Code 07974</b></h1>

Zillow Home Value Index (ZHVI): A smoothed, seasonally adjusted measure of the typical home value and market changes across a given region and housing type. It reflects the typical value for homes in the 35th to 65th percentile range.

[Zillow Data Source](https://www.zillow.com/research/data/)




```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

plt.style.use("bmh")

# Read the CSV file
df = pd.read_csv('https://files.zillowstatic.com/research/public_csvs/zhvi/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')

# Extract the data for zip code 07974
df = df[df['RegionName'] == 7974]

# Select the columns with dates
df = df.loc[:, '2000-01-31':]

# Transpose the data
df = df.T

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200

# Plot the price data
ax = df.plot()

# Add a title
ax.set_title('Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_Prices.pdf')
files.download('07974_Prices.pdf')

# Show the plot
plt.show()

# Compute the moving averages
df_ma3 = df.rolling(3).mean()
df_ma6 = df.rolling(6).mean()
df_ma12 = df.rolling(12).mean()
df_ma24 = df.rolling(24).mean()

# Plot the 3 Month MA data
ax = df_ma3.plot()

# Add a title
ax.set_title('3-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_3_Month_MA.pdf')
files.download('07974_3_Month_MA.pdf')

# Show the plot
plt.show()

# Compute the difference between the original data and the moving average
df_diff3 = df - df_ma3

# Plot the data
ax = df_diff3.plot()

# Add a title
ax.set_title('Difference Between Original Data and 3-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_3_Month_MA_Diff.pdf')
files.download('07974_3_Month_MA_Diff.pdf')

# Show the plot
plt.show()

# Plot the 6 Month MA data
ax = df_ma6.plot()

# Add a title
ax.set_title('6-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_6_Month_MA.pdf')
files.download('07974_6_Month_MA.pdf')

# Show the plot
plt.show()

# Compute the difference between the original data and the moving average
df_diff6 = df - df_ma6

# Plot the data
ax = df_diff6.plot()

# Add a title
ax.set_title('Difference Between Original Data and 6-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_6_Month_MA_Diff.pdf')
files.download('07974_6_Month_MA_Diff.pdf')

# Show the plot
plt.show()

# Plot the 12 Month MA data
ax = df_ma12.plot()

# Add a title
ax.set_title('12-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_12_Month_MA.pdf')
files.download('07974_12_Month_MA.pdf')

# Show the plot
plt.show()

# Compute the difference between the original data and the moving average
df_diff12 = df - df_ma12

# Plot the data
ax = df_diff12.plot()

# Add a title
ax.set_title('Difference Between Original Data and 12-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_12_Month_MA_Diff.pdf')
files.download('07974_12_Month_MA_Diff.pdf')

# Show the plot
plt.show()

# Plot the 24 Month MA data
ax = df_ma24.plot()

# Add a title
ax.set_title('24-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_24_Month_MA.pdf')
files.download('07974_24_Month_MA.pdf')

# Show the plot
plt.show()

# Compute the difference between the original data and the moving average
df_diff24 = df - df_ma24

# Plot the data
ax = df_diff24.plot()

# Add a title
ax.set_title('Difference Between Original Data and 24-Month Moving Average of Housing Prices for Zip Code 07974')

# Add axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Add a legend
ax.legend(['Price'])

plt.savefig('07974_24_Month_MA_Diff.pdf')
files.download('07974_24_Month_MA_Diff.pdf')

# Show the plot
plt.show()
```
