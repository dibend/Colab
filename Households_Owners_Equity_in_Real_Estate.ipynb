```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("bmh")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200

# Load the data from the URL
df = pd.read_csv("https://fred.stlouisfed.org/graph/fredgraph.csv?bgcolor=%23e1e9f0&chart_type=line&drp=0&fo=open%20sans&graph_bgcolor=%23ffffff&height=450&mode=fred&recession_bars=on&txtcolor=%23444444&ts=12&tts=12&width=1168&nt=0&thu=0&trc=0&show_legend=yes&show_axis_titles=yes&show_tooltip=yes&id=OEHRENWBSHNO&scale=left&cosd=1952-01-01&coed=2022-07-01&line_color=%234572a7&link_values=false&line_style=solid&mark_type=none&mw=3&lw=2&ost=-99999&oet=99999&mma=0&fml=a&fq=Quarterly%2C%20End%20of%20Period&fam=avg&fgst=lin&fgsnd=2020-02-01&line_index=1&transformation=lin&vintage_date=2023-01-16&revision_date=2023-01-16&nd=1945-10-01")

# Plot the data 
plt.plot(df["DATE"], df["OEHRENWBSHNO"], label="Households; Owners' Equity in Real Estate, Level in Billions of Dollars", color="green")
plt.fill_between(df["DATE"], df["OEHRENWBSHNO"], alpha=.5, color='green')
plt.xticks(ticks=range(0, len(df["DATE"]), 10), labels=df["DATE"][::10], rotation=90)
plt.xlabel("Date")
plt.ylabel("Level in Billions of Dollars")
plt.title("Households; Owners' Equity in Real Estate")
plt.legend()
plt.show()
```
