```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("bmh")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200

# Load data from Github Gist
df = pd.read_csv('https://gist.githubusercontent.com/dibend/272b0ab55f9c24ed77791c07e27e82ea/raw/f44d9588eb9b7a196ba73aa61fefc2069216d1a2/Russian_Personnel_Losses.csv')

# Plot the data
df.plot(x='date', y='personnel')

plt.xticks(ticks=range(0, len(df['date']), 7), labels=df['date'][::7], rotation=90)
plt.xlabel("Date")
plt.ylabel("Russian Personnel Losses")
plt.title("Russian Personnel Losses Over Time")
# Show the plot
plt.show()
```
